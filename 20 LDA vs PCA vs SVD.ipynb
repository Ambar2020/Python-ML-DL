```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
pca = PCA(n_components=2)
```

```python
X = pca.fit_transform(iris.data)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(X[:,0],X[:,1], c=iris.target)
plt.title('PCA')
```

```python
lda = LinearDiscriminantAnalysis(n_components=2)
```

```python
lda.fit(iris.data, iris.target)
```

```python
X = lda.transform(iris.data)
```

```python
plt.scatter(X[:,0],X[:,1], c=iris.target)
plt.title('LDA')
```

```python
svd = TruncatedSVD(n_components=2)
```

```python
X = svd.fit_transform(iris.data)
```

```python
plt.scatter(X[:,0],X[:,1], c=iris.target)
plt.title('SVD')
```

```python
svd.components_
```

```python
fig,axes = plt.subplots(1,3)

X = pca.fit_transform(iris.data)
axes[0].scatter(X[:,0],X[:,1], c=iris.target)

Y = lda.transform(iris.data)
axes[1].scatter(Y[:,0],Y[:,1], c=iris.target)

X = svd.fit_transform(iris.data)
axes[2].scatter(X[:,0],X[:,1], c=iris.target)
```
