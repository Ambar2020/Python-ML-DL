```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
X = iris.data[:,:2]
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(n_jobs=-1, eps=.3, min_samples=10)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(X[:,0], X[:,1])
```

```python
db.fit(X)
```

```python
db.labels_
```

```python
plt.scatter(X[:,0], X[:,1], c=db.labels_)
```
