```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPool2D
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.io import imread,imshow
```

```python
imshow('rg.jpg')
```

```python
imread('rg.jpg').shape
```

```python
img = imread('rg.jpg')
```

```python
model = Sequential()
model.add(Conv2D(2,3,3,input_shape=img.shape))
```

```python
model.summary()
```

```python
o = model.predict(img.reshape(1,720,1280,3))
```

```python
imshow(o.reshape(718,1278,2)[:,:,0])
```

```python
imshow(o.reshape(718,1278,2)[:,:,1])
```

```python
model = Sequential()
model.add(Conv2D(2,3,3,input_shape=img.shape))
model.add(MaxPool2D(2,2))
model.add(Conv2D(2,3,3, activation='relu'))
```

```python
o = model.predict(img.reshape(1,720,1280,3))
```
