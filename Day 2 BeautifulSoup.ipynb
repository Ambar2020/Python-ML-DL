```python
import requests

response = requests.get("https://www.magicbricks.com/property-for-sale/residential-real-estate?proptype=Multistorey-Apartment,Builder-Floor-Apartment,Penthouse,Studio-Apartment&cityName=Bangalore")

print(response.status_code)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content,"html.parser")
```

```python
print(soup.prettify())
```

```python
price = soup.find("div",attrs={"class":"m-srp-card__price"})
print(price.text)
```

```python
prices = soup.find_all("div",attrs={"class":"m-srp-card__price"})
for price in prices:
    print(price.text)
```

```python
card = soup.find("div",attrs={"class":"m-srp-card__container"})

title = card.find("a",attrs={"class":"m-srp-card__title"})
price = card.find("div",attrs={"class":"m-srp-card__price"})
carpet_area = card.find("div",attrs={"class":"m-srp-card__summary__info"})

new_title = title.text.strip("\n").replace("\n"," ")

# print(new_title)
# print(carpet_area.text)
# print(price.text)

flat_deatils = "{} {} {}".format(new_title,carpet_area.text,price.text)
print(flat_deatils)
```

```python
cards = soup.find_all("div",attrs={"class":"m-srp-card__container"})

for card in cards:
    title = card.find("a",attrs={"class":"m-srp-card__title"})
    price = card.find("div",attrs={"class":"m-srp-card__price"})
    carpet_area = card.find("div",attrs={"class":"m-srp-card__summary__info"})
    
    new_title = title.text.strip("\n").replace("\n"," ")
    flat_deatils = "{} {} {}".format(new_title,carpet_area.text,price.text)
    print(flat_deatils)
```

```python
cards = soup.find_all("div",attrs={"class":"m-srp-card__container"})

for card in cards:
    title = card.find("a",attrs={"class":"m-srp-card__title"})
    price = card.find("div",attrs={"class":"m-srp-card__price"})
    carpet_area = card.find("div",attrs={"class":"m-srp-card__summary__info"})
    
    new_title = title.text.strip("\n").replace("\n"," ")
    
    link = title.get("href")
    response2 = requests.get(link)
    soup2 = BeautifulSoup(response2.content,"html.parser")
#     print(response2.status_code)
    super_area = soup2.find("span",attrs={"id":"coveredAreaDisplay"})
#     print(super_area)
    if super_area:
        final_super_area = super_area.text
    else:
        final_super_area = None
    
    flat_deatils = "{} {} {} {}".format(new_title,carpet_area.text,final_super_area,price.text)
    print(flat_deatils)
```

```python
import csv
cards = soup.find_all("div",attrs={"class":"m-srp-card__container"})
with open("cards2.csv","a+",encoding="UTF-8") as fp:
    writer = csv.writer(fp,lineterminator="\n")
    writer.writerow(['Title','CarpetArea','SuperArea','Price'])
    for card in cards:
        title = card.find("a",attrs={"class":"m-srp-card__title"})
        price = card.find("div",attrs={"class":"m-srp-card__price"})
        carpet_area = card.find("div",attrs={"class":"m-srp-card__summary__info"})

        new_title = title.text.strip("\n").replace("\n"," ")

        link = title.get("href")
        response2 = requests.get(link)
        soup2 = BeautifulSoup(response2.content,"html.parser")
    
        super_area = soup2.find("span",attrs={"id":"coveredAreaDisplay"})
        if super_area:
            final_super_area = super_area.text
        else:
            final_super_area = None
        writer.writerow([new_title,carpet_area.text,final_super_area,price.text])
#         flat_deatils = "{} {} {} {}".format(new_title,carpet_area.text,final_super_area,price.text)
#         print(flat_deatils)
fp.close()
```
