```python
from sklearn.pipeline import Pipeline
from sklearn.decomposition import TruncatedSVD
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler, SMOTE
from imblearn.pipeline import make_pipeline as make_pipeline_imb
from imblearn.metrics import classification_report_imbalanced
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier, GradientBoostingClassifier
```

```python
class ModelCreationValidation(object):
    def __init__(self,models):
        self.pipelines = []
        self.create_pipelines(models)
    
    def create_pipelines(self,models):
        for model in models:
            self.pipelines.append(('smote'+model[0],make_pipeline_imb(SMOTE(),TruncatedSVD(n_components=10),model[1])))
            self.pipelines.append(('OverSample'+model[0],make_pipeline_imb(RandomOverSampler(),TruncatedSVD(n_components=10),model[1])))
            
    def fit(self,X,Y):
        for name,pipeline in self.pipelines:
            pipeline.fit(X,Y)
            

    def grid_fit(self,X,Y,params):
        for name,pipeline in self.pipelines:
            grid = GridSearchCV(pipeline,param_grid=params,cv=5, n_jobs=-1)
            grid.fit(X,Y)
            print (name, grid.best_score_)
            
```

```python
pipelines = ModelCreationValidation([('lr',LogisticRegression()),('rf',RandomForestClassifier()),('ab',AdaBoostClassifier())])
```
