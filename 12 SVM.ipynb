```python
from sklearn.datasets import make_circles
```

```python
X,Y = make_circles(n_samples=10000, noise=.08)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(X[:,0],X[:,1], c=Y, s=10, alpha=.5)
```

```python
from sklearn.svm import SVC
```

```python
params = {
    'svc__kernel':['linear','poly','rbf'],
    'svc__gamma':[0.1,1,10],
    'svc__C':[1,10,100,200]
}
```

```python
from sklearn.pipeline import Pipeline
from sklearn.grid_search import GridSearchCV
```

```python
pipe = Pipeline([
    ('svc',SVC())
])
```

```python
grid = GridSearchCV(pipe, params, cv= 5, n_jobs=-1)
```

```python
grid.fit(X,Y)
```

```python
grid.best_params_
```

```python
grid.best_score_
```
