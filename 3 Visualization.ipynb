```python
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/zekelabs/machine-learning-for-beginners/master/data/house_rental_data.csv', index_col='Unnamed: 0')
```

```python
col = data.columns
cnt = 0
fig,axes = plt.subplots(2,3)
for i in range(2):
    for j in range(3):
        axes[i,j].hist(data[col[cnt]])
        cnt += 1
```
